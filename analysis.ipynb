```python
# for other dataset

if df.isnull().sum().any()==0:
    print("no missing values found")
else:
    df.fillna(0)
    print("missing Value filled with 0")
```

```python
df.duplicated().sum()
```

```python
if df.duplicated().sum() ==0:
    print("no duplicate row found")
else:
    df.drop_duplicates()
    print("Duplicate row dropped")
```

```python
cat_columns = df.select_dtypes(include="object").columns
print("Categorical columns", cat_columns) 
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop = 'first', sparse_output = False)

encoded = encoder.fit_transform(df[cat_columns])

encoded_df = pd.DataFrame(encoded, columns  = encoder.get_feature_names_out(cat_columns),
                          index = df.index)
encoded_df.head()
```

```python
df = pd.concat(
    [df.drop(cat_columns, axis=1), encoded_df],
    axis=1
)

df.head()
```

```python
df.columns
```

```python
#Task 3 — Model Building
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import (mean_absolute_error,mean_squared_error,r2_score)
```

```python
x = df.drop("price",axis = 1)
y = df["price"]
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.20, random_state = 42)
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
lr_predict = linear_model.predict(x_test)
lr_predict[:10]
```

```python
lr_mae = mean_absolute_error(y_test, lr_predict)

lr_rmse = np.sqrt(mean_squared_error(y_test, lr_predict))

lr_r2 = r2_score(y_test, lr_predict)

print(f"Mae is: {lr_mae:,.2f} ,\nRMSE is: {lr_rmse:,.2f} ,\nR2 Score is: {lr_r2:.4f}") 
```

```python
rf_model = RandomForestRegressor(n_estimators = 200, random_state = 42)

rf_model.fit(x_train,y_train)

rf_predict = rf_model.predict(x_test)

rf_predict[:10]
```

```python
rf_mae = mean_absolute_error(y_test, rf_predict)

rf_rmse = np.sqrt(mean_squared_error(y_test, rf_predict))

rf_r2 = r2_score(y_test, rf_predict)

print(f"Mae is: {rf_mae:,.2f} ,\nRMSE is: {rf_rmse:,.2f} ,\nR2 Score is: {rf_r2:.4f}") 
```

```python
#Now we will compare both
```

```python
comp = pd.DataFrame({
    "Model":["linear Regression","Random Forest"],
     "MAE":[lr_mae, rf_mae],
     "RMSE":[lr_rmse,rf_rmse],
     "R2 Score":[lr_r2, rf_r2]

     })

comp  = comp.round(2)
comp #comparison between both models
```

```python
prediction_df = pd.DataFrame({
                              "Actual Price": y_test.values,
                              "Linear Prediction": lr_predict,
                               "Random Forest Prediction": rf_predict
})
prediction_df = prediction_df.round(2)
prediction_df.head(10)
```

```python
best_model = comp.loc[comp["R2 Score"].idxmax(), "Model"]

print(f"Best Performing Model : {best_model}")
```
