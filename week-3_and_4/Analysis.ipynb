```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings

warnings.filterwarnings("ignore")
```

```python
train_data = Monthly_sales[:-3]

test_data = Monthly_sales[-3:]
```

```python
sarima_model = SARIMAX(train_data, order=(1,0,1), seasonal_order=(1,1,1,12),
                      enforce_stationarity=False,enforce_invertibility=False)
```

```python
sarima_result = sarima_model.fit()   #model will learn
```

```python
sarima_result.summary()
```

```python
sarima_forecast = sarima_result.get_forecast(steps = 3)

sarima_prediction = (sarima_forecast.predicted_mean)
sarima_prediction.round(3)
```

```python
sarima_confidence= sarima_forecast.conf_int()
sarima_confidence.round(3)
```

```python
plt.figure(figsize=(10,6))

plt.plot(test_data,label="ACtual sales")

plt.plot(sarima_prediction,label = "SARIMA Forecast")

plt.title("SARIMA Actual vs Forecast Sales", fontsize = 16)

plt.xlabel("Date", fontsize= 14)
plt.ylabel("sales", fontsize = 14)
plt.tight_layout()
plt.legend()

plt.savefig("charts/sarima_forecast.png",dpi=300, bbox_inches="tight")
```

```python
def xgb_forecast(data):            # creating a common function for all sales

    monthly_data = data.resample('M',on='Order Date')['Sales'].sum().reset_index()

    monthly_data['Lag_1'] = (monthly_data['Sales'].shift(1) )

    monthly_data['Lag_2'] = (monthly_data['Sales'].shift(2))

    monthly_data['Lag_3'] = (monthly_data['Sales'].shift(3))

    monthly_data['Rolling_mean'] = (monthly_data['Sales'].rolling(3).mean())

    monthly_data['month'] = (monthly_data['Order Date'].dt.month)

    monthly_data['quarter'] = (monthly_data['Order Date'].dt.quarter)

    monthly_data['season'] = (monthly_data['month'].apply(season))

    monthly_data = monthly_data.dropna()

    X = monthly_data.drop(['Sales','Order Date'],axis=1)

    y = monthly_data['Sales']

    model = xgb_pipeline

    model.fit(X,y)

    prediction = model.predict(X.tail(3))

    return prediction
```

```python
furniture_forecast = xgb_forecast(sales[sales['Category']=='Furniture'])

technology_forecast = xgb_forecast(sales[sales['Category']=='Technology'])

office_forecast = xgb_forecast(sales[sales['Category']=='Office Supplies'])

west_forecast = xgb_forecast(sales[sales['Region']=='West'])

east_forecast = xgb_forecast(sales[sales['Region']=='East'])

```

```python
# Task 5 - Anomaly Detection in Sales Data

from sklearn.ensemble import IsolationForest
```

```python
weekly_sales = sales.resample('W', on = 'Order Date')['Sales'].sum().reset_index()

weekly_sales.head()
```

```python
isolation_model = IsolationForest(contamination = 0.05, random_state = 42)
```

```python
weekly_sales['Anomaly'] = isolation_model.fit_predict(weekly_sales[['Sales']])
weekly_sales['Anomaly'].value_counts()
```

```python
anomaly_points = weekly_sales[weekly_sales['Anomaly'] == -1]
anomaly_points
```

```python
plt.figure(figsize=(12,5))


plt.plot(weekly_sales['Order Date'],
         weekly_sales['Sales'],
         label="Weekly Sales")

plt.scatter(anomaly_points['Order Date'],
           anomaly_points['Sales'],
            color = 'red',
            marker = 'X',
            s = 100,
           label="Anomaly")

plt.title("Sales Anomaly Detection using Isolation Forest")

plt.xlabel("Date")

plt.ylabel("Sales")

plt.legend()

plt.savefig("charts/anomaly_detection.png",dpi=300, bbox_inches="tight")
```

```python
anomaly_points[['Order Date','Sales']]
```

```python
anomaly_points = anomaly_points.copy()

average_sales = weekly_sales['Sales'].mean()


anomaly_points['Possible Reason'] = np.where(anomaly_points['Sales'] > average_sales,

    "Possible festive sale, promotional campaign or seasonal demand spike.",

    "Possible supply shortage, reduced customer demand or operational issue.")


anomaly_points[['Order Date',
                'Sales',
                'Possible Reason']]
```

```python
#Z-score Detection
```

```python
weekly_sales["Rolling_Mean"] = (weekly_sales["Sales"].rolling(4).mean())

weekly_sales["Rolling_STD"] = (weekly_sales["Sales"].rolling(4).std())

weekly_sales["Z_score"] = (weekly_sales["Sales"]- weekly_sales["Rolling_Mean"]) / weekly_sales["Rolling_STD"]

weekly_sales.head()
```

```python
print("Minimum Z-score:", round(weekly_sales["Z_score"].min(),3))
print("Maximum Z-score:", round(weekly_sales["Z_score"].max(),3))
```

```python
z_anomalies = weekly_sales[abs(weekly_sales['Z_score'])>2].copy()
z_anomalies
```

```python
weekly_sales["Z_score"].describe()
weekly_sales["Z_score"].sort_values()
```

```python
# Z-Score Anomaly Detection

# The Z-Score method was applied using the required threshold of ±2 standard deviations from the rolling mean.

# No weekly sales observations exceeded this threshold. This indicates that, according to the statistical rule, the weekly sales remained within the expected range and no extreme outliers were detected.
```

```python
common_dates = set(
    anomaly_points['Order Date']).intersection(set
                                               (z_anomalies['Order Date'])
)
print("Common anomalies:",len(common_dates))

print("Isolation Forest anomalies:",len(anomaly_points))

print("Z-Score anomalies:",len(z_anomalies))
```

```python
# The two methods do not completely agree. Isolation Forest detected 11 anomalous weeks, while the Z-Score method detected none.
    
# This indicates that Isolation Forest is better at identifying complex sales patterns, whereas the Z-Score method only detects statistically extreme values.
    
# For this dataset, Isolation Forest proved to be the more effective anomaly detection method.
```

```python
#Task -6 Product Demand Segmentation using Clustering
```

```python
segment_data = sales.groupby("Sub-Category").agg({"Sales":"sum",
                                                  "Order ID": "count"}).reset_index()

segment_data.head()
```

```python
segment_data["Average_Order_Value"] = (segment_data["Sales"]/ segment_data["Order ID"])
segment_data.head()   #Average Order Value
```

```python
monthly_subcategory = sales.groupby([ "Sub-Category", pd.Grouper(key="Order Date",freq="M")])["Sales"].sum().reset_index()

monthly_subcategory.head()
```

```python
volatility = monthly_subcategory.groupby("Sub-Category")["Sales"].std().reset_index()

volatility.rename(columns={"Sales":"Sales_Volatility"},inplace=True) 

volatility.head()
```

```python
segment_data = segment_data.merge(volatility,on="Sub-Category")

segment_data.head()
```

```python
yearly_sales = sales.groupby(["Sub-Category","Year"])["Sales"].sum().reset_index()

yearly_sales.head()
```

```python
growth_rate = yearly_sales.groupby(
    "Sub-Category"
).apply(lambda x: ((x.iloc[-1]["Sales"]-x.iloc[0]["Sales"])/x.iloc[0]["Sales"]) * 100).reset_index(name="Growth_Rate")

growth_rate.head()
```

```python
segment_data = segment_data.merge(growth_rate,on="Sub-Category")

segment_data.head()
```

```python
X_cluster = segment_data.drop( "Sub-Category",axis=1)

X_cluster.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X_cluster)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):

    kmeans = KMeans( n_clusters=i,random_state=42)

    kmeans.fit( X_scaled)

    wcss.append(kmeans.inertia_)
  
```

```python
  len(wcss)
```

```python
plt.figure(figsize=(8,5))

plt.plot( range(1,11), wcss,marker="o")

plt.title( "Elbow Method")

plt.xlabel("Number of Clusters")

plt.ylabel("WCSS")

plt.savefig("charts/Elbow Method.png",dpi=300, bbox_inches="tight")

```

```python
kmeans = KMeans(n_clusters=4,random_state=42)

segment_data["Cluster"] = kmeans.fit_predict( X_scaled)

segment_data.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)

pca_data = pca.fit_transform(X_scaled)
```

```python
segment_data["PCA1"] = pca_data[:,0]

segment_data["PCA2"] = pca_data[:,1]
```

```python
plt.figure(figsize=(10,6))

plt.scatter(segment_data["PCA1"],
            segment_data["PCA2"],
            c=segment_data["Cluster"],s=100)

for i in range(len(segment_data)):
    plt.text( segment_data["PCA1"][i],
              segment_data["PCA2"][i],
              segment_data["Sub-Category"][i],
              fontsize=8)

plt.title("Product Demand Segmentation using K-Means")

plt.xlabel("Principal Component 1")

plt.ylabel("Principal Component 2")

plt.savefig("charts/product Demand.png",dpi=300, bbox_inches="tight")
```
